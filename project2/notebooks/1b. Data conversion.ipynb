### Imports and preparation

```python
import os
import pandas as pd
```

```python
# Where are we going to work?
data_path = "../data/prelim/preprocess2/out/"
pickle_path = "../data/prelim/work/out.pickle"
out_path = "../data/prelim/work/annotate.txt"
```

```python
df = pd.read_pickle(pickle_path)
df.head()
```

### Conversion
The format for the `keras-frcnn` is `filepath,x1,y1,x2,y2,class_name`. We need to create a text file with this format.

```python
df['file_name'] = data_path + df['file_name']
df.head()
```

```python
col = list(df.columns)
col = [col[0]] + col[2:] + [col[1]]
df = df[col]
df.head()
```

### Saving

```python
df.to_csv(out_path, index=False, sep=',', header=None)
```

### Some exploration
